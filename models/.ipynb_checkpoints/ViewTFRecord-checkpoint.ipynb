```python
import pdb
TFRecord_files = ['lx_median_2009-11_uganda_dhslocs_ee_export.tfrecord.gz']#,
#    'ls_median_2011-13_AfroR5_1ee_export.tfrecord.gz']

#TFRecord_files = ['ls_median_2014-2015_afroappproj_225x225_8ee_export.tfrecord.gz']

band_names = ['BLUE', 'GREEN', "RED", "NIR", "SWIR1", "SWIR2"]
#band_names = ['RED', 'BLUE', 'GREEN', 'SWIR1', 'SWIR2', 'NIR', 'NIGHTLIGHTS']
def analyze_tfrecord(tfrecord_path):
    # Create an iterator over the TFRecords file. The iterator yields
    # the binary representations of Example messages as strings.
    options = tf.python_io.TFRecordOptions(tf.python_io.TFRecordCompressionType.GZIP)
    iterator = tf.python_io.tf_record_iterator(TFRecords_file, options=options)
    
    images_count = 0
    image_size = 225
    
    sums = {}
    sq_sums = {}
    for band_name in band_names:
        sums[band_name] = 0
        sq_sums[band_name] = 0
    
    for record_str in iterator:
        ex = tf.train.Example.FromString(record_str)  # parse into an actual Example message
        feature_map = ex.features.feature  # get the mapping from feature name strings to Feature
        
        for band_name in band_names:
            sums[band_name] += np.sum(feature_map[band_name].float_list.value)
            sq_sums[band_name] += np.sum(np.array(feature_map[band_name].float_list.value) ** 2)
        
        images_count += 1
        if images_count % 2 == 0:
            print("Finished {} images...".format(images_count))
    
    for band_name in band_names:
        num_total_pixels = images_count * image_size**2
        mean = sums[band_name] / float(num_total_pixels)
        std = (sq_sums[band_name] - sums[band_name]) / float(num_total_pixels)
        print("Band {} - mean: {:0.5f}, std: {:0.5f}".format(band_name, mean, std))
    
    return images_count, sums, sq_sums

def analyze_tfrecord_batch(tfrecord_paths):
    batch_size = 100
    
    dataset = tf.data.TFRecordDataset(tfrecord_paths, compression_type="GZIP")
    dataset = dataset.map(_parse_function)  # Parse the record into tensors.
    dataset = dataset.batch(batch_size)
    iterator = dataset.make_one_shot_iterator()
    next_element = iterator.get_next()
    
    images_count = 0
    image_size = 255
    
    import os
    os.environ['CUDA_VISIBLE_DEVICES'] = ''
    sess = tf.Session()
    
    mins = {}
    maxes = {}
    sums = {}
    sq_sums = {}
    count = 1
    for band_name in band_names:
        mins[band_name] = np.inf
        maxes[band_name] = 0
        sums[band_name] = 0
        sq_sums[band_name] = 0
    
    try:
        while True:
            batch = sess.run(next_element)
            print(batch.keys())
            pdb.set_trace()
            count += 1
            for band_name in band_names:
                mins[band_name] = min(mins[band_name], np.min(batch['min'][band_name]))
                maxes[band_name] = max(maxes[band_name], np.max(batch['max'][band_name]))
                sums[band_name] += np.sum(batch['sum'][band_name])
                sq_sums[band_name] += np.sum(batch['sum_sq'][band_name])

            images_count += batch_size
            if images_count % 100 == 0:
                print("Finished {} images...".format(images_count))
    except:        
         pass
    
    print(images_count)
    for band_name in band_names:
        num_total_pixels = images_count * image_size**2
        print(images_count)
        mean = sums[band_name] / float(num_total_pixels)
        std = np.sqrt(sq_sums[band_name]/float(num_total_pixels) - (sums[band_name]/float(num_total_pixels))**2)
        print("Band {} - mean: {:0.5f}, std: {:0.5f}, min: {:0.5f}, max: {:0.5f}".format(
            band_name, mean, std, mins[band_name], maxes[band_name]))
#         print("Band {} - sum: {:0.5f}, sq_sum: {:0.5f}".format(
#             band_name, sums[band_name], sq_sums[band_name]))
    
    return images_count, mins, maxes, sums, sq_sums
    

def _parse_function(example_proto):
    features = {
        band_name : tf.FixedLenFeature(shape=[255**2], dtype=tf.float32)
        for band_name in band_names
    }
    parsed_features = tf.parse_single_example(example_proto, features)
    
    for band_name in band_names:
        parsed_features[band_name] = tf.cast(parsed_features[band_name], tf.float64)
    
    new_feats = {
        'min': {},
        'max': {},
        'sum': {},
        'sum_sq': {}
    }
    
    for band_name in band_names:
        new_feats['min'][band_name] = tf.reduce_min(parsed_features[band_name])
        new_feats['max'][band_name] = tf.reduce_max(parsed_features[band_name])
        new_feats['sum'][band_name] = tf.reduce_sum(parsed_features[band_name])
        new_feats['sum_sq'][band_name] = tf.reduce_sum(parsed_features[band_name] ** 2)
    
    return new_feats
```

```python
# Images are 500x500 numpy arrays 
def load_model_cnn():
    np.random.seed(1234)

    model = Sequential()
    model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(255,255,3)))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.25))
    model.add(Conv2D(64, (3, 3), activation='relu'))
    model.add(Flatten())
    model.add(Dense(128, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, kernel_initializer='normal'))

    model.compile(loss='mean_squared_error', optimizer='adam')
    print(model.summary())
    return model
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from textwrap import wrap

fig = plt.figure(figsize=[13, 7])

def set_wrapped_title(ax, title):
    wrapped_title = '\n'.join(wrap(title, 25))
    ax.set_title(wrapped_title)

ax = plt.subplot(2, 4, 1)
set_wrapped_title(ax, 'RGB (Bands 3,2,1) - 15m')
ax.imshow(ms_image_float32[:, :, [2,1,0]])
ax.axis('off')

ax = plt.subplot(2, 4, 2)
set_wrapped_title(ax, 'Near Infared (NIR) (Band 4) - 30m')
ax.imshow(ms_image_float32[:, :, 3], cmap='gray')
ax.axis('off')

ax = plt.subplot(2, 4, 3)
set_wrapped_title(ax, 'Shortwave Infrared 1 (SWIR1) (Band 5) - 30m')
ax.imshow(ms_image_float32[:, :, 4], cmap='gray')
ax.axis('off')

ax = plt.subplot(2, 4, 4)
set_wrapped_title(ax, 'Thermal 1 (Band 6 VCID 1) - 60m')
ax.imshow(ms_image_float32[:, :, 7], cmap='gray')
ax.axis('off')

ax = plt.subplot(2, 4, 5)
set_wrapped_title(ax, 'Thermal 2 (Band 6 VCID 2) - 60m')
ax.imshow(ms_image_float32[:, :, 8], cmap='gray')
ax.axis('off')

ax = plt.subplot(2, 4, 6)
set_wrapped_title(ax, 'Shortwave Infrared 2 (SWIR2) (Band 7) - 30m')
ax.imshow(ms_image_float32[:, :, 5], cmap='gray')
ax.axis('off')

ax = plt.subplot(2, 4, 7)
set_wrapped_title(ax, 'Panchromatic (Band 8) - 15m')
ax.imshow(ms_image_float32[:, :, 6], cmap='gray')
ax.axis('off')

fig.tight_layout(pad=0)
plt.savefig('bands.png', bbox_inches='tight', transparent=True, pad_inches=0)
```
